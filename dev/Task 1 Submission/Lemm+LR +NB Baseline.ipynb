```python
!pip install -U imbalanced-learn

import nltk
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
import scipy as scipy
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from imblearn.keras import BalancedBatchGenerator
from imblearn.under_sampling import NearMiss
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import f1_score, precision_score, recall_score

nltk.download('punkt')
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

train_file = "train.data.txt"
dev_file = "dev.data.txt"
test_file = "test.data.txt"


train_data = pd.read_csv('./%s.csv'%train_file,keep_default_na=False)
dev_data = pd.read_csv('./%s.csv'%dev_file,keep_default_na=False)
test_data = pd.read_csv('./%s.csv'%test_file,keep_default_na=False)


def tokenize_tweet(string_data:str):
  stem = PorterStemmer()
  wordnet_lemmatizer = WordNetLemmatizer()
  tokenized = nltk.RegexpTokenizer('\w+')
  data = string_data.replace('\n', ' ')
  #remove links
  data = re.sub('http\S+', '', data)
  # print(data)
  data = data.lower()
  data = tokenized.tokenize(data)
  data = [i for i in data if i not in stopwords]
  data = [stem.stem(word) for word in data]
  data = [wordnet_lemmatizer.lemmatize(word) for word in data]
  print(data)
  return data

def preprocess(pd_data):
  data = pd_data.copy()
  data['main_tweet'] = data['main_tweet'].fillna('')
  data.replace('', np.nan, inplace=False)
  data.dropna(subset=['main_tweet'], inplace=True)
  data['main_tweet'] = data['main_tweet'].apply(lambda x: x.replace('\n', ' ')) 
  # data['main_tweet'] = data['main_tweet'].apply(lambda x: tokenize_tweet(x))
  return data
le = LabelEncoder()


train_data = preprocess(train_data)
dev_data = preprocess(dev_data)


# tfidf_vec = TfidfVectorizer(tokenizer=tokenize_tweet)

# train = tfidf_vec.fit_transform(train_data['main_tweet'])

# train = scipy.sparse.csr_matrix.todense(train)
# dev = tfidf_vec.transform(dev_data['main_tweet'])
# dev = scipy.sparse.csr_matrix.todense(dev)

# test = tfidf_vec.transform(test_data['main_tweet'])
# test = scipy.sparse.csr_matrix.todense(test)



# train_label = le.fit_transform(train_data['label'])


# dev_label = le.transform(dev_data['label'])

# training_generator = BalancedBatchGenerator(train, train_label, sampler=NearMiss(), batch_size=10, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv_vec = CountVectorizer(tokenizer=tokenize_tweet)

train = cv_vec.fit_transform(train_data['main_tweet'])
train = scipy.sparse.csr_matrix.todense(train)

dev = cv_vec.transform(dev_data['main_tweet'])
dev = scipy.sparse.csr_matrix.todense(dev)

test = cv_vec.transform(test_data['main_tweet'])
test = scipy.sparse.csr_matrix.todense(test)

train_label = le.fit_transform(train_data['label'])


dev_label = le.transform(dev_data['label'])

```

```python
LRclassifier = LogisticRegression()
LRclassifier.fit(train, train_label)

NBclassifier = MultinomialNB()
NBclassifier.fit(train, train_label)
```

```python

lr_pred_dev_prob = LRclassifier.predict_proba(dev)
nb_pred_dev_prob = NBclassifier.predict_proba(dev)

lr_predict = LRclassifier.predict(dev)
nb_predict = NBclassifier.predict(dev)
```

```python

print("Logistic Regression")
print("precision:", precision_score(dev_label, lr_predict))
print("recall:", recall_score(dev_label, lr_predict))
print("f1:", f1_score(dev_label, lr_predict))

print("\n")
print("Naives Bayes Model")
print("precision:", precision_score(dev_label, nb_predict))
print("recall:", recall_score(dev_label, nb_predict))
print("f1:", f1_score(dev_label, nb_predict))
```

```python
#Predict LR and output submission
prediction = LRclassifier.predict(test)
prediction = (prediction > 0.5).astype("int32")
prediction = np.ndarray.flatten(prediction)
pd.DataFrame({"Predicted":  prediction}).to_csv('Lemm+LR-submit.csv', index_label="Id")
```
